# Emotion Detection

Importing libraries

```python
#importing libraries
import matplotlib.pyplot as plt  #to test our model at final
from glob import glob               #to create a list of all files in directory
import cv2                          #to read images and perform operations on images
import random                       #to generate anything in random
import os

#no errors while plotting graphs
%matplotlib inline

#importing preprocessing functions for ML model
from tensorflow.keras.preprocessing.image import ImageDataGenerator #generates data so , ML model understands

from tensorflow.keras.layers import Dense, Input, Dropout, Flatten, Conv2D
# Dense - linear function
# Input - decides input size
# Dropout - using it, so it uses not only training dataset , uses also external inputs
# Flatten - Changes a matrix to oneD array
# Conv2D - conversation layer, read images and learn from it

from tensorflow.keras.layers import BatchNormalization , Activation , MaxPooling2D
# BatchNormalization - used to do normalization of data
# Activation - to help the network learn complex patterns in the data
# MaxPooling2D - takes the maximum value within that window

from tensorflow.keras.models import Model , Sequential
# Model - create or define answer of model
# Sequential - type of tensorflow model

from tensorflow.keras.optimizers import Adam # Adam - type of optimizer
from tensorflow.keras.callbacks import ModelCheckpoint , ReduceLROnPlateau
# ModelCheckpoint - store data on each and every data on each end , at a every point of iteration
# ReduceROnPlateau - Decrease learning rate or no learning rate

from tensorflow.keras.utils import plot_model # plot_model - to plot training losses
from IPython .display import SVG , Image # used to display image in jupyter notebook
import tensorflow as tf
```

**Preparing data for training **

```python
img_size = 48
batch_size = 64
datagen_train = ImageDataGenerator()
train_generator = datagen_train.flow_from_directory("train/" , target_size = (img_size,img_size) ,color_mode = "grayscale" , 
                                                    batch_size = batch_size , class_mode = "categorical" ,
                                                   shuffle = True)

datagen_validation= ImageDataGenerator()
validation_generator = datagen_train.flow_from_directory("test/" , target_size = (img_size,img_size) ,color_mode = "grayscale" , 
                                                    batch_size = batch_size , class_mode = "categorical" ,
                                                   shuffle = True)
```

*Defining model*

```python
def Convolution(input_tensor , filters , kernel_size):
    x = Conv2D(filters = filters , kernel_size = kernel_size , padding = "same")(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size = (2,2))(x)
    x = Dropout(0.25)(x)
    
    return x
```

```python
def Dense_f(input_tensor , nodes):
    x = Dense(nodes)(input_tensor)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(0.25)(x)
    
    return x
```

```python
def model_fer(input_shape):
    inputs = Input(input_shape)
    conv_1 = Convolution(inputs,32,(3,3))
    conv_2 = Convolution(conv_1,64,(5,5))
    conv_3 = Convolution(conv_2,128,(3,3))

    flatten = Flatten()(conv_3)
    
    dense_1 = Dense_f(flatten,256)

    output  = Dense(7, activation="softmax")(dense_1)
    model  =  Model(inputs=[inputs],outputs = [output])
    
    model.compile(loss=['categorical_crossentropy'] , optimizer = 'adam' , metrics =["accuracy"])

    return model
```

```python
model = model_fer((48,48,1))
model.summary()
```

*Initializing Model*

```python
epochs = 15
steps_per_epoch = train_generator.n // train_generator.batch_size
validation_steps = validation_generator.n // validation_generator.batch_size

checkpoint = ModelCheckpoint("model_weights.h5", monitor  ='val_accuracy', save_weights_only = True, mode = 'max', verbose =1)
callbacks = [checkpoint]
```

*Training the model*

```python
history = model.fit(
x = train_generator,
steps_per_epoch = steps_per_epoch,
epochs = epochs,
validation_data = validation_generator,
validation_steps = validation_steps,
callbacks = callbacks)
```

*Model Evaluation*

```python
model.evaluate(validation_generator)
```

*plotting loss*

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.legend(['Train','Validation'],loc = 'upper left')
plt.subplots_adjust(top=1.0,bottom=0.0,right =0.95,left=0.0,hspace=0.25,wspace=0.35)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.legend(['Train','Validation'],loc = 'upper left')
plt.subplots_adjust(top=1.0,bottom=0.0,right =0.95,left=0.0,hspace=0.25,wspace=0.35)
```
